```python
links = dataTwitter["Link"]
menciones = []
driver = webdriver.Edge(executable_path='C:/Users/Afreedom/Practicas/msedgedriver.exe')

xpath_mencion="/html/body/div/div/div/div[2]/main/div/div/div/div/div/div[2]/div/section/div/div/div[1]/div/div/article/div/div/div/div[3]/div[1]/div/div"
xpath_retweet="/html/body/div/div/div/div[2]/main/div/div/div/div/div/div[2]/div/section/div/div/div[2]/div/div/article/div/div/div/div[3]/div[2]"
for l in links:
    driver.get(str(l))
    time.sleep(5)
    try:
        elem = driver.find_element_by_xpath(xpath_mencion) 
        texto = elem.text
    except NoSuchElementException:
        try:
            elem = driver.find_element_by_xpath(xpath_retweet) 
            texto = elem.text
        except NoSuchElementException:
            texto = "No encontrado"
    
    texto = texto.replace('\n', ' ')
    menciones.append(texto)
    print(texto)
driver.close()
```
