```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv("dataWhatsapp.csv", sep = ';')
data["fecha"] = ""
data["hora"] = ""

timestamp = data["timestamp"]

fecha = []
hora = []
```

```python
for i in timestamp:
    fechaFormat = datetime.datetime.fromtimestamp(i / 1000).strftime('%Y-%m-%d')
    horaFormat = datetime.datetime.fromtimestamp(i / 1000).strftime('%H:%M:%S')
    fecha.append(fechaFormat)
    hora.append(horaFormat)
```
